```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv("logistic.csv")
df.head()
```

```python

X = df[['male', 'age', 'education', 'currentSmoker', 'cigsPerDay', 'BPMeds',
          'prevalentStroke', 'prevalentHyp', 'diabetes', 'totChol', 'sysBP',
          'diaBP', 'BMI', 'heartRate', 'glucose']]
Y = df['TenYearCHD']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
X_train
```

```python
Y_train
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
X_train_imputed = imputer.fit_transform(X_train)
X_test_imputed = imputer.transform(X_test)
model = LogisticRegression(max_iter=10000)
model.fit(X_train_imputed, Y_train)
```

```python
Y_pred = model.predict(X_test_imputed)
Y_pred
```

```python
accuracy = accuracy_score(Y_test, Y_pred)
accuracy
```

```python
Y_prob = model.predict_proba(X_test_imputed)[:, 1]
plt.scatter(X_test['age'], Y_prob, alpha=0.5, c=Y_test ,cmap='coolwarm')
plt.xlabel('Age')
plt.ylabel('Probability of Heart Disease (TenYearCHD = 1)')
plt.title('Scatter Plot: Age vs. Probability of Heart Disease')
plt.colorbar(label='Actual Heart Disease (0 = No, 1 = Yes)')
plt.show()
```

```python
import numpy as np
def predict_heart_disease(model, user_input):

    user_data = np.array(user_input).reshape(1, -1)
    prediction = model.predict(user_data)
    prediction_prob = model.predict_proba(user_data)[:, 1]
    return prediction[0], prediction_prob[0]

# Example user input
user_input = [1, 55, 1, 1, 0, 1, 1, 1, 1, 300, 140, 850, 28, 75, 85]

# Make a prediction
predicted_class, predicted_prob = predict_heart_disease(model, user_input)

# Display the result
if predicted_class == 1:
    print(f"The model predicts that the person has a high risk of heart disease with a probability of {predicted_prob:.2f}.")
else:
    print(f"The model predicts that the person has a low risk of heart disease with a probability of {predicted_prob:.2f}.")
```

```python
import pickle

# Save the model to a file
with open('logistic_regression_model.pkl', 'wb') as file:
    pickle.dump(model, file)

print("Model saved successfully as 'logistic_regression_model.pkl'")
```
